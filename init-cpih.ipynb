# hpi index

```python
df = pd.read_csv("./UK-HPI-full-file-2023-09.csv")
```

```python
df['Date'] = df.Date.apply(lambda x: pd.to_datetime(x, format="%d/%m/%Y"))
```

```python
df.RegionName.unique()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
_l = df.columns[df.columns.str.endswith("Index")][0:5].tolist()
for _l in _l:
    sns.lineplot(df, x="Date", y=_l, label=_l, ax=ax)
```

```python
idf = df[['Date', 'RegionName'] + df.columns[df.columns.str.endswith("Index")][0:5].tolist()].sort_values(by=["Date", "RegionName"]).dropna()
```

```python
idf.RegionName.value_counts()
```

```python
sns.lineplot(idf[idf.RegionName == "England"], x="Date", y="Index", label="England")
sns.lineplot(idf[idf.RegionName == "Wales"], x="Date", y="Index", label="Wales")
```
